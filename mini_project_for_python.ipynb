```python
menu = {
    "pizza": 40,
    "pasta": 40,
    "burger": 60,
    "salad": 70,
    "dosa": 65,
    "idli": 65,
    "coffee": 80,
    "lemon rice": 60
}

print("Welcome to the Restaurant!")
print("\n     MENU     ")
for item, price in menu.items():
    print(f"{item.title():<12} : Rs {price}")


order_total = 0
order_list = []

while True:
    item = input("Enter the item you want to order (or type 'done' to finish): ").lower()

    if item == "done":
        break
    elif item in menu:
        quantity = input(f"How many {item}(s) do you want? ")

        if quantity.isdigit():
            quantity = int(quantity)
            order_total += menu[item] * quantity
            order_list.append((item, quantity, menu[item] * quantity))
            print(f" {quantity} x {item} added to your order.")
        else:
            print(" Please enter a valid number for quantity.")
    else:
        print(f" Sorry, {item} is not available on the menu.")


print("\n    BILL    ")
for item, qty, price in order_list:
    print(f"{item.title()} (x{qty}) : Rs {price}")
print(f" Total Amount to Pay: Rs {order_total}")

```
